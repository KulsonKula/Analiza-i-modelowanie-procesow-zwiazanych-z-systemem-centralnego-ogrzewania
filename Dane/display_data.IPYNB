```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
DATASET=pd.read_csv('data_nowe/inputData/ewh_profile[12].csv')
```

```python
DATASET
```

```python
y=DATASET.Summer_Water_Consumption.iloc[0:1440]
x=np.arange(0,24,24/1440) 
```

```python
y=y.iloc[0:1440]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6))


axs[0].plot(x, y)
axs[0].set_title('Całość dnia')
axs[0].set_xlim(0, 24)
axs[0].set_xlabel('Godzina')
axs[0].set_ylabel('Przepływ')

axs[1].scatter(x, y)
axs[1].set_title('Zbliżenie na godzine 7:20')
axs[1].set_xlim(7.2, 7.5)
axs[1].set_xlabel('Godzina')
axs[1].set_ylabel('Przepływ')

axs[2].scatter(x, y)
axs[2].set_title('Zbliżenie na godzine 15:00')
axs[2].set_xlim(14.8, 15.3)
axs[2].set_xlabel('Godzina')
axs[2].set_ylabel('Przepływ')
```
